## Check GPU Availability

## Configuration: Adjust Hyperparameters Here

```python
# EDIT THESE VALUES TO EXPERIMENT
NUM_EPOCHS = 50
LEARNING_RATE = 0.001
BATCH_SIZE = 32
PATIENCE = 10
CHECKPOINT_NAME = "colab_experiment_v1"  # Change this for each experiment

print(f"Training Configuration:")
print(f"  Epochs: {NUM_EPOCHS}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Patience: {PATIENCE}")
print(f"  Checkpoint Name: {CHECKPOINT_NAME}")
```

## Run Training

```python
from data_loader import load_cifar10
from model import SimpleCNN
from training import TrainingConfig, train_model, plot_training_curves

# Load data
print("Loading CIFAR10 dataset...")
train_loader, val_loader, test_loader, classes = load_cifar10()

# Create model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model = SimpleCNN()
model = model.to(device)

# Create config with custom hyperparameters
config = TrainingConfig(checkpoint_name=CHECKPOINT_NAME)
config.num_epochs = NUM_EPOCHS
config.learning_rate = LEARNING_RATE
config.batch_size = BATCH_SIZE
config.patience = PATIENCE
config.device = device

# Train model
print("\nStarting training...")
model, history = train_model(model, train_loader, val_loader, config)

# Plot curves
print("\nPlotting training curves...")
plot_training_curves(history, save_path=f"checkpoints/{CHECKPOINT_NAME}_curves.png")

# Summary
print("\n" + "="*50)
print("TRAINING COMPLETE")
print("="*50)
print(f"Best validation accuracy: {max(history['val_accuracy']):.2f}%")
print(f"Final validation accuracy: {history['val_accuracy'][-1]:.2f}%")
print(f"Checkpoint: checkpoints/{CHECKPOINT_NAME}.pt")
print("="*50)
```

## Download Results to Local Machine

```python
from google.colab import files

print("Downloading checkpoint and results...\n")

# Download model checkpoint
files.download(f'checkpoints/{CHECKPOINT_NAME}.pt')
print(f"[OK] Downloaded checkpoint: {CHECKPOINT_NAME}.pt")

# Download training history
files.download(f'checkpoints/{CHECKPOINT_NAME}_history.json')
print(f"[OK] Downloaded history: {CHECKPOINT_NAME}_history.json")

# Download training curves
files.download(f'checkpoints/{CHECKPOINT_NAME}_curves.png')
print(f"[OK] Downloaded curves: {CHECKPOINT_NAME}_curves.png")

print("\nAll files downloaded! Move them to your local 'checkpoints/' folder.")
```
